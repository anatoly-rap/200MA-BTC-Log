```python
import matplotlib.pyplot as plt;
import matplotlib as mpl;
from matplotlib.ticker import ScalarFormatter
import pandas as pd;
```

```python
mpl.rcParams['figure.figsize'] = (20,20);
df = pd.read_csv(r'C:\....\*.csv',index_col = 0, parse_dates = True, usecols = ['Date','Close']);
print(df.head());        
```

```python
newCol = df['Close'].rolling(window = 200, min_periods=1).mean();
df["200 Day MA"] = newCol;
print(df.head());

fig, axes = plt.subplots(nrows=2, ncols=1);

df.plot(ax=axes[0], logy = True, grid = True);
axes[0].set(title = "log Scale BTC Price & EMA", ylabel = "United States Dollars");
axes[0].yaxis.set_major_formatter(ScalarFormatter())


newCol.plot(ax=axes[1], ylim = (0,100000), grid = True);
axes[1].set(title = "BTC Linear Price History", ylabel = "United States Dollars");

```
